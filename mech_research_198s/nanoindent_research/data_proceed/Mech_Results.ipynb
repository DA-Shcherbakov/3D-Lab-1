```python
def extract_label():
    global name_data_Er_list
    list_V = []
    list_alg = []
    list_slice =[]
    list_ser = []
    for name in name_data_Er_list:
        len_name = len(name)
        str_list = list(name)
        
        ##list_V
        str_tmp = []
        tmp = []
        i = 0
        j = 0
        while name[i] !='_':
            i+=1
        i+=1
        while name[i+j] !="m":
            str_tmp +=str_list[i+j]
            j+=1

        tmp = int(''.join(str_tmp))
    
        list_V.append(tmp)
        
        ##
        ##list_alg
        i += j
        j=0
        str_tmp = []
        #print(i,j)
        #print(name[i])
        while name[i] !='_':
            i+=1
        i+=1
        #print(i,name[i])
        while name[i+j] !="s":
            str_tmp +=str_list[i+j]
            j+=1
        str_tmp +=str_list[i+j]
        tmp = ''.join(str_tmp)
        list_alg.append(tmp)
        
        ##
        ##list_slice
        i += j
        j=0
        str_tmp = []
        #print(i,j)
        #print(name[i])
        while name[i] !='_':
            i+=1
        i+=1
        #print(i,name[i])
        while name[i+j] !="_":
            str_tmp +=str_list[i+j]
            j+=1
        
        tmp = ''.join(str_tmp)
        list_slice.append(tmp)
        
        ##list_ser
        tmp = str_list[-1]
        list_ser.append(tmp)
        ##
        
    return list_V, list_alg, list_slice, list_ser
```

```python
def extract_label_DF(DF):
    
    len_DF = len(DF)
    
    list_LP = []
    list_V = []
    list_alg = []
    list_slicing = []
    for k in range(len_DF):
        
        name = list(DF['Sample'][k])
        len_name = len(name)
       # print(name,DF['Sample'][k])
        tmp = []
        str_tmp =[]
        i = 0
        j = 0
        
        
        ##
        ##list_LP
        
        while name[i] !='_':
            i+=1
        i+=1
        while name[i+j] !="_":
            str_tmp +=name[i+j]
            j+=1

        tmp = float(''.join(str_tmp))
        tmp = float(tmp/2)
    
        list_LP.append(tmp)
        
        ##
        ##list_alg
        i += j +1
        j=0
        str_tmp = []
        #print(i,j)
        #print(name[i])
        #print(i,name[i])
        while name[i+j] !="s":
            str_tmp +=name[i+j]
            j+=1
        str_tmp +=name[i+j]
        tmp = ''.join(str_tmp)
        list_alg.append(tmp)
        
        ##
        ## list_V
        i += j
        j=0
        str_tmp = []
        #print(i,name[i])
        while name[i] !='_':
            i+=1
        i+=1
        while name[i+j] !="m":
            str_tmp +=name[i+j]
            j+=1

        tmp = int(''.join(str_tmp))
    
        list_V.append(tmp)
        ##
        ##
        ##list_slice
        i += j
        j=0
        str_tmp = []
        #print(i,j)
        #print(name[i])
        while name[i] !='(' and i < (len_name-1):
            i+=1
            
        i+=1
        #print(i,name[i])
        if i == len_name:
            list_slicing.append('True')
        else:
            
            while name[i+j] !=")":
                str_tmp +=name[i+j]
                j+=1

            tmp = ''.join(str_tmp)
            list_slicing.append(tmp)
        
    
    DF['LP,mW'] = list_LP
    DF['V,mkmps'] = list_V
    DF['Algoritm'] = list_alg
    DF['Slicing'] = list_slicing
    return DF
```
