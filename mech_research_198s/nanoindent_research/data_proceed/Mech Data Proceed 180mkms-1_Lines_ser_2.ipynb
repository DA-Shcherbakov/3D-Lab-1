```python
def reader_of_LP(sample_list):
    list_LP = []
    for name in sample_list:
        len_name = len(name)
        str_list = list(name)
        str_tmp = []
        i = 0
        j = 0
        while name[i] !='P':
            i+=1
        i+=1
        while name[i+j] !="_":
            str_tmp +=str_list[i+j]
            j+=1
            
        tmp = int(''.join(str_tmp))
        if tmp < 30:
            tmp = float(tmp/2)
        else:
            tmp = float(tmp/10/2)
            
        list_LP.append(tmp)
    return list_LP
```
